```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/data_csv/World_Energy_By_Country_And_Region_1965_to_2023.csv")
df
```

2. cleaning data

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df[~df['Country'].str.contains('Total', case=False, na=False)]
```

```python
df = df.head(len(df) - 4)
```

```python
df = df.dropna(subset=['Country'])
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df
```

```python
year_columns = df.columns[1:]
```

```python
for col in year_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dtypes
```

```python
df['Average_Energy_Usage'] = df[year_columns].mean(axis=1, skipna=True)
```

```python
print(df[['Country', 'Average_Energy_Usage']].head())
```

```python
df.info()
```

3. explanatory data(EDA)

4. label encoder

```python
from sklearn.preprocessing import LabelEncoder

C_encoder = LabelEncoder()

df['Country'] = C_encoder.fit_transform(df['Country'])
```

```python
df.sample(5)
```

5. preparation data

```python
features = df.columns.drop('Country')  # Mengambil semua kolom kecuali 'Country'
x = df[features]  # Data fitur
y = df['Country']  # Data target
x.shape, y.shape  # Menampilkan bentuk dari x dan y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
y_test.shape
```

6. Training & evaluation

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')  # atau 'median', 'most_frequent', dll.
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print("akurasi model regresi linear berganda adalah ", score)
```
